```python
# xに座標点、yにクラスラベルが代入される。
from sklearn.datasets import make_blobs
x,y = make_blobs(n_samples = 300,
                n_features = 2,#特徴量ふたつ
                centers = 3,
                random_state = 0)
```

```python
# 散布図の作成
import matplotlib.pyplot as plt
%matplotlib inline

plt.scatter(x[:,0],
           x[:,1],
           c = 'blue',
           marker = 'o',
           s =50)
```

```python
# 予測
from sklearn.cluster import KMeans
km = KMeans(n_clusters=5)
km.fit(x)
km_predict= km.predict(x)
print(km_predict)
```

```python
km_predict.shape
```

```python
# 描写
plt.scatter(x[km_predict == 0,0],#座標点
            x[km_predict == 0,1],#クラスラベル
            s =50,#プロットのサイズ
            c = 'green',#プロットの色
            label = 'cluster1')#ラベル名
plt.scatter(x[km_predict == 1,0],
            x[km_predict == 1,1],
            s =50,
            c = 'purple',
            label = 'cluster2')
plt.scatter(x[km_predict == 2,0],
            x[km_predict == 2,1],
            s =50,
            c = 'red',
            label = 'cluster3')
plt.scatter(x[km_predict == 3,0],
            x[km_predict == 3,1],
            s =50,
            c = 'blue',
            label = 'cluster4')
plt.scatter(x[km_predict == 4,0],
            x[km_predict == 4,1],
            s =50,
            c = 'orange',
            label = 'cluster5')
plt.scatter(km.cluster_centers_[:,0],
           km.cluster_centers_[:,1],
           marker = '*',
           s = 250,
           c = 'yellow',
           label = 'centroid')
```

```python
# エルボー法
print('Sum of distance', km.inertia_)
```

```python
# 可視化する
distances=[]
for i in range(1,11):
    km = KMeans(n_clusters = i)
    km.fit(x)
    distances.append(km.inertia_)#SSEを配列に追加
    
# グラフの描写
plt.plot(range(1,11),distances,marker='.')
plt.xlabel('Number of Clusters')
plt.ylabel('Sum of Distances')
```
