```python
# CSVファイルの読み込み
import pandas as pd
pokemon = pd.read_csv('pokemon.csv')
pokemon.head()
```

```python
# データをNumpyの配列に変換,いらない特徴を省く。
import numpy as np
x = np.array(pokemon)
x = x[:,4:10]
```

```python
# pcaクラスの読み込み
from sklearn.decomposition import PCA
# 主成分数を2に設定したインスタンスを生成
pca = PCA(n_components=2)
pca.fit(x)
# データについて学習したPCAで主成分を抽出する
x_pca = pca.transform(x)
print(x_pca)
```

```python
x_pca.shape
```

データが二次元の配列に。一列目が、第一主成分。二列目が、第二主成分。

```python
# データを散布図に描写
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
x = x_pca[:,0]#カラムは[x,y]のふたパターンあり、それが８０１レコードある。つまり、xには、第一主成分の０カラムを、yには第二主成分の１カラムを代入しているのである。
y = x_pca[:,1]
fig, ax = plt.subplots(figsize=(8,8))#オブジェクト指向インターフェイス
plt.scatter(x, y)
plt.xlabel('1st principal component') #第一主成分 
plt.ylabel('2nd principal component') #第二主成分 
plt.show()
```

```python
pca.explained_variance_ratio_
```

```python
# 因子負荷量の確認
pca.components_
```

```python
# 異なる散布図の作成
attr = ['attack','defense','sp_attack','sp_defense','speed','hp']
fig, ax = plt.subplots(figsize = (8,8))
plt.scatter(x,y)
for i in range(pca.components_.shape[1]):
    x1=pca.components_[0,i]*100
    y1=pca.components_[1,i]*100
    # .arrow( X座標 , Y座標 , X長さ , Y長さ , head_width=ヘッド幅 , head_length=ヘッド長 )
    ax.arrow(0,0,x1,y1,head_width = 5,head_length = 10,facecolor='black')
    plt.text(x1+15,y1,attr[i])
plt.show
```
